```python
#!/usr/bin/env python
# -*- coding: utf-8 -*-
# !pip install jieba


import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt
# plt.get_cachedir()
plt.rcParams['font.sans-serif']=['SimHei']

from PIL import Image
import numpy as np
from lxml import etree
from nltk.tokenize import word_tokenize
# pip install --user -U nltk
```

```python
def remove_stop_words(f):
    stop_words=['The']
    for stop_word in stop_words:
        f=f.replace(stop_word,'')
    return f
    
```

```python
#生成词云功能
def create_word_cloud(f):
    print("根据词频，开始生成词云：")
    f=remove_stop_words(f)
    cut_text=word_tokenize(f)
    cut_text=" ".join(cut_text)
    wc=WordCloud(max_words=40,width=2000,height=1200)
    wordcloud=wc.generate(cut_text)
    wordcloud.to_file("wordcloud.jpg")
    plt.imshow(wordcloud)
    plt.axis("off")
    plt.show()
    
    
    
    
    
   
```

```python
# 数据加载
data=pd.read_csv('./Market_Basket_Optimisation.csv',header=None)
transactions=[]
#再加字典作分析
item_count={}
for i in range(0,data.shape[0]):
    temp=[]
    for j in range(0,20):
        item=str(data.values[i,j])
        if item!='nan':
            temp.append(item)
            #增加字典判断是否在字典中，
            if item in item_count:
                item_count[item]+=1
            else:
                item_count[item]=1
            
    transactions.append(temp)
# 生成词云
all_words=' '.join('%s' % item for item  in transactions)
# 喂入词云
create_word_cloud(all_words)

#输出TOP10
All_item=sorted(item_count.items(),key=lambda x:x[1],reverse=True)
print ('TOP10: ')
print (All_item[:11])
```
